## Apply anonymization on multiple articles StiripeSurse


```python
data_news = pd.read_csv(f"{PATH}/Research/Datasets/NonS/non-sarcastic_articles_final_step2.csv")
```

```python
data_news.head(n=1)
```

```python
nlp = spacy.load('ro_core_news_sm')
spacy.tokens.token.Token.set_extension('text_anonymized', default='')
```

```python
def add_article_anonymized(article:str)->int:
  article_anonymized = None
  doc = nlp(article)
  ents = [e.text for e in doc.ents]
  person = [e for e in doc.ents if e.label_ == 'PERSON']
  locations_gpe = [e for e in doc.ents if e.label_ == 'GPE']
  locations_non_gpe = [e for e in doc.ents if e.label_ == 'LOC']
  org = [e for e in doc.ents if e.label_ == 'ORGANIZATION']
  event = [e for e in doc.ents if e.label_ == 'EVENT']
  for tok in doc: 
    tok._.text_anonymized = tok.text
  for ent in person:
      ent[0]._.text_anonymized = "$NE$"
      for i in range(1, len(ent)):
          ent[i]._.text_anonymized = ''
  for ent in locations_gpe:
      ent[0]._.text_anonymized = "$NE$"
      for i in range(1, len(ent)):
          ent[i]._.text_anonymized = ''
  for ent in locations_non_gpe:
      ent[0]._.text_anonymized = "$NE$"
      for i in range(1, len(ent)):
          ent[i]._.text_anonymized = ''
  for ent in org:
      ent[0]._.text_anonymized = "$NE$"
      for i in range(1, len(ent)):
          ent[i]._.text_anonymized = ''
  for ent in event:
      ent[0]._.text_anonymized = "$NE$"
      for i in range(1, len(ent)):
          ent[i]._.text_anonymized = ''
  article_anonymized = "".join([tok._.text_anonymized + (" " if tok.whitespace_ else "")
         for tok in doc if tok._.text_anonymized])
  
  return article_anonymized
```

```python
data_news['article_anonymized'] = data_news.article_processed.swifter.apply(add_article_anonymized)
```

```python
data_news['title_anonymized'] = data_news.title_processed.swifter.apply(add_article_anonymized)
```

```python
data_news.to_csv(f"{PATH}/Research/Datasets/NonS/non-sarcastic_articles_anonymized_step2.csv", index=False)
```
