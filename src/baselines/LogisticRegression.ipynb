# LogisticRegression 

```python
article_train, article_test, label_train, label_test = train_test_split(article, label, test_size=0.1, random_state=42) 
```

```python
article_vectorizer = CountVectorizer()
article_vectorizer.fit(article_train)
Xlr_train = article_vectorizer.transform(article_train)
Xlr_test  = article_vectorizer.transform(article_test)
```

```python
label_train
```

```python
Xlr_train
```

Now the data is almost prepared for fitting in the model, lets perform the logistic regression model building and fitting in the data. 

```python
LRmodel = LogisticRegression()
LRmodel.fit(Xlr_train, label_train)
```

```python
score = LRmodel.score(Xlr_test, label_test)
print("Accuracy:", score)
```

```python
predictions=LRmodel.predict(Xlr_test)
print(classification_report(label_test,predictions))
```

Here we have seen the text classification model with very basic levels. 

```python
confusion_matrix(label_test, predictions) 
```
